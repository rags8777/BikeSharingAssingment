```python
# From the RFE Model, the following variables : yr, temp,hum,windspeed, season(3,4) and weathersit influenced cnt
# Looking into the stats model with variables

from sklearn.model_selection import train_test_split

# Creating the training and test set with a split 70/30

df_train, df_test = train_test_split(bike_sharing, train_size = 0.7, test_size = 0.3, random_state = 100)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated

plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
# cnt is highly correlated to yr, temp, atemp, season(summer, fall), weather(light wind)
# Chosing temp as the first variable
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
# Fit a regression line through the training data using statsmodels (Chosing temp as the first variable)
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train[['temp']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Check the parameters obtained
lr.params
```

```python
# Print a summary of the linear regression model obtained
print(lr.summary())
```

```python
# The R2 obtained is 0.414 -Adding the second varaible atemp

# Add a constant
X_train_lm = sm.add_constant(X_train[['temp','atemp']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
```

```python
# Check the summary
print(lr.summary())
```

```python
# The R2 obtained is 0.419 with temp / atemp
# Adding the year variable
# Add a constant
X_train_lm = sm.add_constant(X_train[['temp','atemp','yr']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
```

```python
# Check the summary
print(lr.summary())
```

```python
# The R2 obtained is 0.696 with temp / atemp / yr
# Adding the season and weathersit varaibles
# Add a constant
X_train.columns = X_train.columns.astype(str)
X_train_lm = sm.add_constant(X_train[['temp','atemp','yr','3','4']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
```

```python
# Check the summary
print(lr.summary())
```

```python
# R2 values is at 0.765
# p value is greater than 0.05 for atemp, season variable (3)
# Adding variable workingday
X_train_lm = sm.add_constant(X_train[['temp','atemp','yr','3','4','workingday']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
```

```python
# Check the summary
print(lr.summary())
```

```python
# Calculate the VIFs again for the new model
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_lm.columns
vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(X_train_lm.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Removing atemp and season variable (3) from the model
X_train_lm = sm.add_constant(X_train[['temp','yr','4','workingday']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
```

```python
# Check the summary
print(lr.summary())
```

```python
# Adding hum and windspeed and season(2,3) to the model
# The R-squared is nearly 0.826
# p values are less than 0.05
X_train_lm = sm.add_constant(X_train[['temp','yr','2','4','3','workingday','hum','windspeed']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
lr.params
# Check the summary
print(lr.summary())
```

```python
# Calculate the VIFs again for the new model
# Check for the VIF values of the feature variables. 
# The VIF for the following variables was less than 5 - 'temp','yr',season(spring,summer,fall),weathersit(2,3,4),'workingday','hum','windspeed'
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_lm.columns
vif['VIF'] = [variance_inflation_factor(X_train_lm.values, i) for i in range(X_train_lm.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python

# As per the model variables that influence ctn are : 'temp','yr',season(spring,summer,fall),weathersit(2,3,4),'workingday','hum','windspeed'
# VIF for the the variables 'temp','yr',season(spring,summer,fall),weathersit(2,3,4),'workingday','hum','windspeed' is less than 5

# Check if error terms are normally distributed
y_train_price = lr.predict(X_train_lm)

# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```
